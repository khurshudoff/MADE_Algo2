```python
INF = 1844674407370955161500
 
class Edge:
    def __init__(self, _from, _to, _fl, _cap, w, idx):
        self.fr = _from
        self.to = _to
        self.fl = _fl
        self.cp = _cap
        self.idx = idx
        self.w = w
            
    def __repr__(self):
        return "({} --> {}: {}/{})".format(self.fr, self.to, self.fl, self.cp)

def bfs():
    way = [-1] * len(g)
    queue = [1]
    way[1] = 0
    while len(queue) > 0 and way[n] == -1:
        s = queue.pop(0) 
 
        for edg in g[s]: 
            if way[edg.to] == -1 and (edg.cp - edg.fl) != 0: 
                queue.append(edg.to) 
                way[edg.to] = s
 
    if way[n] == -1:
        return []
    res = []
    cur = n
    while cur != 0:
        res.append(cur)
        cur = way[cur]
    return list(reversed(res))
 
def bellFord():
    d = [[INF, -1] for _ in range(len(g))]
    d[1] = [0, -1]
    for it in range(len(g)):
        for v in g:
            for e in v:
                if e.cp == e.fl:
                    continue
                if d[e.to][0] > d[e.fr][0] + e.w:
                    d[e.to] = [d[e.fr][0] + e.w, e]
    
    cur = d[n][1]
    way = [cur]
    while cur.fr != 1:
        cur = d[cur.fr][1]
        way.append(cur)
    
    return d[n][0], list(reversed(way))
 
def get_c_min(way):
    return min([e.cp - e.fl for e in way])
 
def inc(way, v):
    for e in way:
        e.fl += v
        e.ed.fl -= v
    

n, m = list(map(int, input().split()))
 
arr = []
for _ in range(m):
    inp = list(map(int, input().split()))
    arr.append(inp)
    
sp = [0 for _ in range(m)]
    
edges = []
    
g = [[] for _ in range(n + 1)]
    
for idx, (fr, to, cp, w) in enumerate(arr):
    g[fr].append(Edge(fr, to, 0, cp, w, idx))
    g[to].append(Edge(to, fr, 0, 0, -w, -idx))
    g[fr][-1].ed = g[to][-1]
    g[to][-1].ed = g[fr][-1]
    
res = 0
    
while len(bfs()) > 0:
    min_w, way = bellFord()
    c_min = get_c_min(way)
    inc(way, c_min)
    res += min_w * c_min
    
print(res)
```

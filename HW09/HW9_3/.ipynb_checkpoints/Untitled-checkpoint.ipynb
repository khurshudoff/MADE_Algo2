```python
INF = 2147483647
 
class Edge:
    def __init__(self, _from, _to, _fl, _cap, w, idx):
        self.fr = _from
        self.to = _to
        self.fl = _fl
        self.cp = _cap
        self.idx = idx
        self.w = w
            
    def __repr__(self):
        return "({} --> {}: {}/{})".format(self.fr, self.to, self.fl, self.cp)

def bfs():
    way = [-1] * len(g)
    queue = [1]
    way[1] = 0
    while len(queue) > 0 and way[n] == -1:
        s = queue.pop(0) 
 
        for edg in g[s]: 
            if way[edg.to] == -1 and (edg.cp - edg.fl) != 0: 
                queue.append(edg.to) 
                way[edg.to] = s
 
    if way[n] == -1:
        return False
    return True
 
def bellFord():
    d = [[INF, -1] for _ in range(len(g))]
    d[1] = [0, -1]
    queue = [1]
    is_in_queue = [0] * len(g)
    is_in_queue[1] = 1
    while len(queue) > 0:
        cur = queue[0]
        del queue[0]
        is_in_queue[cur] = 0
        for e in g[cur]:
            if e.cp != e.fl:
                if d[e.to][0] > d[e.fr][0] + e.w:
                    d[e.to] = [d[e.fr][0] + e.w, e]
                    if not is_in_queue[e.to]:
                        queue.append(e.to)
                        is_in_queue[e.to] = 1
    
    cur = d[n][1]
    way = [cur]
    while cur.fr != 1:
        cur = d[cur.fr][1]
        way.append(cur)
    
    return d[n][0], list(reversed(way))
 
def get_c_min(way):
    return min([e.cp - e.fl for e in way])
 
def inc(way, v):
    for e in way:
        e.fl += v
        e.ed.fl -= v
        if e.idx > 0:
            sp[abs(e.idx)] += v
        else:
            sp[abs(e.idx)] -= v
            
def dist(u, v):
    return abs(inp_N[u][0] - inp_M[v][0]) + abs(inp_N[u][1] - inp_M[v][1]) + 1

N, M = list(map(int, input().split()))
inp_N = []
for _ in range(N):
    inp_N.append(list(map(int, input().split()))) # x, y, amm

inp_M = []

for _ in range(M):
    inp_M.append(list(map(int, input().split()))) # x, y, amm

sp_inp = []
for _ in range(N):
    sp_inp += list(map(int, input().split()))

arr = []
    
# make edges from stock
for v in range(N):
    arr.append([1, v + 2, inp_N[v][2], 0])
    
# make edges between rows and columns
for u in range(N):
    for v in range(M):
        arr.append([u + 2, v + N + 2, INF, dist(u, v)])

for u in range(N + 2, N + M + 2):
    arr.append([u, N + M + 2, inp_M[u - N - 2][2], 0])
    
m = len(arr)
    
sp = [0 for _ in range(m)]
    
edges = []

n = N + M + 2
    
g = [[] for _ in range(n + 1)]
    
for idx, (fr, to, cp, w) in enumerate(arr):
    g[fr].append(Edge(fr, to, 0, cp, w, idx))
    g[to].append(Edge(to, fr, 0, 0, -w, -idx))
    g[fr][-1].ed = g[to][-1]
    g[to][-1].ed = g[fr][-1]
    
    
while bfs():
    min_w, way = bellFord()
    c_min = get_c_min(way)
    inc(way, c_min)

sp = sp[N:-M]

res_inp = 0
for i in range(N):
    for j in range(M):
        res_inp += sp_inp[M * i + j] * dist(i, j)

res = 0    
for i in range(N):
    for j in range(M):
        res += sp[M * i + j] * dist(i, j)
        
        
if res == res_inp:
    print('OPTIMAL')
else:
    print('SUBOPTIMAL')

    for i in range(0, len(sp), M):
        print(' '.join(list(map(str, sp[i:i+M]))))
```

```python
def sign(v):
    return 1 if v > 0 else -1

def bfs(s, t):
    way = [-1] * len(g)
    queue = [s]
    way[s] = 0
    while len(queue) > 0 and way[t] == -1:
        cur = queue.pop(0) 

        for edg in g[cur]: 
            if way[edg.to] == -1 and (edg.cp - edg.fl) != 0: 
                queue.append(edg.to) 
                way[edg.to] = cur

    if way[t] == -1:
        return []
    res = []
    cur = t
    while cur != 0:
        res.append(cur)
        cur = way[cur]
    return list(reversed(res))

def get_c_min(way):
    c_min = INF
    for i in range(len(way) - 1):
        for e in g[way[i]]:
            if e.to == way[i + 1]:
                c_min = min(c_min, abs(e.cp - e.fl))
                break

    return c_min

def inc(way, v):
    for i in range(len(way) - 1):
        for j, e in enumerate(g[way[i]]):
            if e.to == way[i + 1]:
                e.fl += abs(v) * sign(e.cp)
                sp[abs(e.idx)] += abs(v) * sign(e.cp)
                e.ed.fl -= abs(v) * sign(e.cp)
                if e.fl == e.cp:
                    del g[way[i]][j]
                break

INF = 5000 * 10000

class Edge:
    def __init__(self, _from, _to, _fl, _cap, idx):
        self.fr = _from
        self.to = _to
        self.fl = _fl
        self.cp = _cap
        self.idx = idx
            
    def __repr__(self):
        return "({} --> {}: {}/{})".format(self.fr, self.to, self.fl, self.cp)
            
def dfs(s, t):
    st = [s]
    way = [(-1, -1) for _ in range(len(g))]
    way[s] = (0, -1)
    while len(st) > 0:
        cur = st[-1]
        del st[-1]
        if cur == t:
            break
        
        for v, idx in g[cur]:
            if way[v][0] != -1:
                continue
            st.append(v)
            way[v] = (cur, idx)
    
    if way[t][0] == -1:
        return [], []
    cur = t
    res_way = []
    to_del_idx = []
    while cur != 0:
        res_way.append(cur)
        to_del_idx.append(way[cur][1])
        cur = way[cur][0]
        
    return list(reversed(res_way)), list(reversed(to_del_idx))[1:]

n, m, s, t = list(map(int, input().split()))
arr = []
for _ in range(m):
    inp = list(map(int, input().split()))
    arr.append(inp)

sp = [0 for _ in range(m + 1)]

g = [[] for _ in range(n + 1)]

for idx, el in enumerate(arr):
    idx += 1
    _from, _to = el
    g[_from].append(Edge(_from, _to, 0, 1, idx))
    g[_to].append(Edge(_to, _from, 0, 0, -idx))
    g[_from][-1].ed = g[_to][-1]
    g[_to][-1].ed = g[_from][-1]
    
res = 0
while True:
    way = bfs(s, t)
    if len(way) == 0:
        break
    c_min = get_c_min(way)
    res += abs(c_min)
    inc(way, c_min)
    
# оставляем только те ребра, по которым течет поток, запускаем дфс с удалением
arr = [el for idx, el in enumerate(arr) if sp[idx + 1]]
    
g = [[] for _ in range(n + 1)]

for idx, (fr, to) in enumerate(arr):
    g[fr].append((to, idx))
    
    
way1, to_del_idx = dfs(s, t)
to_del_idx = set(to_del_idx)

arr = [el for i, el in enumerate(arr) if i not in to_del_idx]

g = [[] for _ in range(n + 1)]

for idx, (fr, to) in enumerate(arr):
    g[fr].append((to, idx)) 
    
way2, _ = dfs(s, t)

if len(way2) == 0:
    print('NO')
else:
    print('YES')
    print(' '.join(list(map(str, way1))))
    print(' '.join(list(map(str, way2))))
```

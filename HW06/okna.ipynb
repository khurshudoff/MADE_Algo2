```python
class Rect:
    def __init__(self, x1, y1, x2, y2):
        self.x1 = x1
        self.y1 = y1
        self.x2 = x2
        self.y2 = y2        
        
def is_in(x, y, r):
    return x >= r.x1 and x <= r.x2 and y >= r.y1 and y <= r.y2

def f2(arr):
    r_arr = []
    p_arr = []
    for x1, y1, x2, y2 in arr:
        r_arr.append(Rect(x1, y1, x2, y2))
        p_arr.append((x1, y1))
        p_arr.append((x2, y2))

    cnt = 0
    p = (0, 0)
    for x in range(2*100):
        for y in range(2*100):
            cur_cnt = 0
            for r in r_arr:
                cur_cnt += is_in(x, y, r)
            if cur_cnt > cnt:
                cnt = cur_cnt
                p = (x, y)

    return (cnt, p[0], p[1])
```

```python
def f1(arr):
    x_arr = []
    y_arr = []

    for x1, y1, x2, y2 in arr:
        x2+=1
        y2+=1
        x_arr.append(x1)
        x_arr.append(x2)

        y_arr.append((y1, x1, x2, 1))
        y_arr.append((y2, x1, x2, -1))

    x_arr = sorted(x_arr)
    y_arr = sorted(y_arr, key=lambda el:(el[0], -el[-1]))


    ans = [0]
    for i in range(1, len(x_arr)):
        l_arr = []
        cnt = 0
        for el in y_arr:
            if el[2] <= x_arr[i - 1] or el[1] >= x_arr[i]:
                continue
            cnt += el[-1]
            if el[-1] == 1:
                l_arr.append(el[1])
            else:
                del l_arr[l_arr.index(el[1])]
            if cnt > ans[0]:
                ans = cnt, max(l_arr), el[0]

    return ans
```

```python
def sign(v):
    return 1 if v > 0 else -1

def bfs():
    way = [-1] * len(g)
    queue = [1]
    way[1] = 0
    while len(queue) > 0 and way[n] == -1:
        s = queue.pop(0) 

        for edg in g[s]: 
            if way[edg.to] == -1 and (edg.cp - edg.fl) != 0: 
                queue.append(edg.to) 
                way[edg.to] = s

    if way[n] == -1:
        return []
    res = []
    cur = n
    while cur != 0:
        res.append(cur)
        cur = way[cur]
    return list(reversed(res))

def get_c_min(way):
    c_min = INF
    for i in range(len(way) - 1):
        for e in g[way[i]]:
            if e.to == way[i + 1]:
                c_min = min(c_min, abs(e.cp - e.fl))
                break

    return c_min

def inc(way, v):
    for i in range(len(way) - 1):
        for j, e in enumerate(g[way[i]]):
            if e.to == way[i + 1]:
                e.fl += abs(v) * sign(e.cp)
                sp[abs(e.idx)] += abs(v) * sign(e.cp)
                e.ed.fl -= abs(v) * sign(e.cp)
                if e.fl == e.cp:
                    del g[way[i]][j]
                break

INF = 5000 * 10000

class Edge:
    def __init__(self, _from, _to, _fl, _cap, idx):
        self.fr = _from
        self.to = _to
        self.fl = _fl
        self.cp = _cap
        self.idx = idx
            
    def __repr__(self):
        return "({} --> {}: {}/{})".format(self.fr, self.to, self.fl, self.cp)
            
n = int(input())
m = int(input())
arr = []
for _ in range(m):
    inp = list(map(int, input().split()))
    arr.append(inp)

sp = [0 for _ in range(m + 1)]

g = [[] for _ in range(n + 1)]

for idx, el in enumerate(arr):
    idx += 1
    _from, _to, _cap = el
    g[_from].append(Edge(_from, _to, 0, _cap, idx))
    g[_to].append(Edge(_to, _from, 0, -_cap, -idx))
    g[_from][-1].ed = g[_to][-1]
    g[_to][-1].ed = g[_from][-1]
    
res = 0
while True:
    way = bfs()
    if len(way) == 0:
        break
    c_min = get_c_min(way)
    res += abs(c_min)
    inc(way, c_min)
    
print(res)
print('\n'.join(list(map(str, sp[1:]))))
```
